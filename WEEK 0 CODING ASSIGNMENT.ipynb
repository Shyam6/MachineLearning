```python
#read meal_info.csv file and printing first 5 rows
df_meal=pd.read_csv('C:\\Users\\tilak\\Desktop\\MyPython\\assignment\\train_GzS76OK\\meal_info.csv')
df_meal.head()
```

```python
#read fulfilment_center_info.csv file and printing first 5 rows
df_center=pd.read_csv('C:\\Users\\tilak\\Desktop\\MyPython\\assignment\\train_GzS76OK\\fulfilment_center_info.csv')
df_center.head()

```

```python
#read train.csv file and printing first 5 rows
df_food=pd.read_csv('C:\\Users\\tilak\\Desktop\\MyPython\\assignment\\train_GzS76OK\\train.csv')
df_food.head()



```

```python
#merging df_meal and df_food
df_mealandfood=pd.merge(df_meal,df_food,on='meal_id')
df_mealandfood
#it is the "df" mentioned according to what you have given in "WEEK 0 CODING ASST" I renamed it as "df_mealandfood"
```

```python
#merging df_center and df_food
df_centerandfood=pd.merge(df_center,df_food,on='center_id')
df_centerandfood
```

```python
table = pd.pivot_table(data=df_mealandfood,index='category',values='num_orders',aggfunc=np.sum)
#pivot table function takes the data from the dataframe we provided as an argument.By the arguments index and values we can make whatever column as a index and we can have multiindexes also.by using aggfunc we can perfome many operations...using numpy functions also line np.sum or doing mean(it is default).
#so basically I wanna say Pivot table function is used to summarize and aggregate data in a dataframe
table
```

```python
table.plot.bar()
#plots bargraph
plt.xticks(rotation=70)
#This xticks function with rotation argument rotates "food item names written to 70 degrees" this xtixks function can do many other things also
plt.title('Most popular food')
#prints title
plt.ylabel('Quantity sold')
#prints ylabel
plt.xlabel('Food item')
#prints xlabel
```

```python
df_mealandfood['revenue']=df_mealandfood['checkout_price']*df_mealandfood['num_orders']
#new revenue column added
df_mealandfood['month']=df_mealandfood['week']//4
#new month column added
df_mealandfood
```

```python
table2=pd.pivot_table(data=df_mealandfood,index='week',values='revenue',aggfunc=np.sum)
table3=pd.pivot_table(data=df_mealandfood,index='month',values='revenue',aggfunc=np.sum)
#again pivot_table func is used . Now it creates a dataframe having week as index and in aggfunc we used np.sum operation
# and in index we wont have any repeating value and therefore now in the dataframe table2 we get index as week and values as 
#sum of all the revenue(only column in table2) in that particular week
#same is the case with table3 and months
```

```python
week=[]
week_order=[]
month=[]
month_order=[]
week=table2.index
week_order=table2.values
x=np.array(week)
y=np.array(week_order)
month=table3.index
month_order=table3.values
p=np.array(month)
q=np.array(month_order)
fig=plt.figure(figsize=(20,5))
#basically we created a space/figure object of dimensions mentioned abv in paranthesis(simply the dimensions of both plots together)
w1=fig.add_subplot(1,2,1)
#here a subplot first argument is row number,second column number,last one is position of the that plot
#when we use figure object then to get subplot we use add_subplot
m1=fig.add_subplot(1,2,2)
w1.plot(x,y)
m1.plot(p,q)
w1.set_title('Weekly income')#title
#to set the title we use set_title 
w1.set_xlabel('Week')#xlabel
w1.set_ylabel('Revenue')#ylabel
m1.set_title('Monthly income')#title
m1.set_xlabel('Month')#xlabel
m1.set_ylabel('Revenue')#ylabel
plt.show()
```
